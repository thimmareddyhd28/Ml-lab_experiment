```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns

def auto_detect_structure(filename='wdbc.data'):
    """
    Automatically detect the structure of the UCI dataset and generate appropriate column names
    """
    print("Auto-detecting dataset structure...")

    # First, load the data to understand its structure
    temp_data = pd.read_csv(filename, header=None)

    print(f"Dataset shape detected: {temp_data.shape}")
    print(f"Number of columns: {temp_data.shape[1]}")
    print(f"Number of rows: {temp_data.shape[0]}")

    # Analyze first few rows to understand structure
    print("\nFirst 3 rows (raw data):")
    print(temp_data.head(3))

    # Detect diagnosis column by looking for M/B values
    diagnosis_col = None
    for col in temp_data.columns:
        unique_vals = temp_data[col].unique()
        if len(unique_vals) == 2 and set(unique_vals) == {'M', 'B'}:
            diagnosis_col = col
            break

    if diagnosis_col is None:
        print("Warning: Could not detect diagnosis column. Assuming column 1 contains diagnosis.")
        diagnosis_col = 1

    print(f"Diagnosis column detected at position: {diagnosis_col}")
    print(f"Diagnosis values: {sorted(temp_data[diagnosis_col].unique())}")

    # Generate column names automatically
    num_cols = temp_data.shape[1]
    column_names = []

    # First column is typically ID
    column_names.append('id')

    # Second column is typically diagnosis
    column_names.append('diagnosis')

    # Remaining columns are features
    # For WDBC dataset, we know there are 30 features (10 base features × 3 statistics each)
    remaining_cols = num_cols - 2

    if remaining_cols == 30:  # WDBC format
        print("Detected WDBC format: 30 features (10 base features × 3 statistics)")

        # Base feature names (common medical measurements)
        base_features = ['radius', 'texture', 'perimeter', 'area', 'smoothness',
                        'compactness', 'concavity', 'concave_points', 'symmetry', 'fractal_dimension']

        # Statistics computed for each base feature
        statistics = ['mean', 'se', 'worst']

        # Generate feature names
        for stat in statistics:
            for feature in base_features:
                column_names.append(f'{feature}_{stat}')
    else:
        # Generic feature naming
        print(f"Generic format detected: {remaining_cols} features")
        for i in range(remaining_cols):
            column_names.append(f'feature_{i+1}')

    print(f"Generated {len(column_names)} column names")
    print(f"Column names: {column_names[:5]} ... {column_names[-5:]}")

    return column_names

def analyze_dataset_automatically(data):
    """
    Automatically analyze the dataset structure and provide insights
    """
    print(f"\n{'='*60}")
    print("AUTOMATIC DATASET ANALYSIS")
    print(f"{'='*60}")

    # Basic information
    print(f"Dataset dimensions: {data.shape[0]} rows × {data.shape[1]} columns")
    print(f"Memory usage: {data.memory_usage(deep=True).sum() / 1024:.1f} KB")

    # Check data types
    print(f"\nData types detected:")
    dtype_counts = data.dtypes.value_counts()
    for dtype, count in dtype_counts.items():
        print(f"  {dtype}: {count} columns")

    # Missing values
    missing_total = data.isnull().sum().sum()
    print(f"\nMissing values: {missing_total}")

    if missing_total > 0:
        missing_by_col = data.isnull().sum()
        missing_cols = missing_by_col[missing_by_col > 0]
        print("Columns with missing values:")
        for col, count in missing_cols.items():
            print(f"  {col}: {count} missing ({count/len(data)*100:.1f}%)")

    # Identify categorical and numerical columns
    categorical_cols = data.select_dtypes(include=['object']).columns.tolist()
    numerical_cols = data.select_dtypes(include=[np.number]).columns.tolist()

    print(f"\nColumn types:")
    print(f"  Categorical columns: {len(categorical_cols)}")
    if categorical_cols:
        print(f"    {categorical_cols}")

    print(f"  Numerical columns: {len(numerical_cols)}")
    if len(numerical_cols) <= 10:
        print(f"    {numerical_cols}")
    else:
        print(f"    {numerical_cols[:5]} ... {numerical_cols[-5:]}")

    # Analyze target variable if 'diagnosis' column exists
    if 'diagnosis' in data.columns:
        print(f"\nTarget variable analysis (diagnosis):")
        diagnosis_counts = data['diagnosis'].value_counts()
        print(f"  Classes found: {list(diagnosis_counts.index)}")

        for diagnosis, count in diagnosis_counts.items():
            percentage = count / len(data) * 100
            diagnosis_name = "Malignant" if diagnosis == "M" else "Benign" if diagnosis == "B" else diagnosis
            print(f"  {diagnosis_name} ({diagnosis}): {count} samples ({percentage:.1f}%)")

        # Check class balance
        balance_ratio = diagnosis_counts.max() / diagnosis_counts.min()
        if balance_ratio > 3:
            print(f"  ⚠️  Class imbalance detected (ratio: {balance_ratio:.1f}:1)")
        else:
            print(f"  ✓ Classes are reasonably balanced (ratio: {balance_ratio:.1f}:1)")

    # Numerical features analysis
    if numerical_cols:
        numeric_data = data[numerical_cols]
        print(f"\nNumerical features statistics:")
        print(f"  Feature count: {len(numerical_cols)}")
        print(f"  Value ranges:")
        print(f"    Minimum: {numeric_data.min().min():.6f}")
        print(f"    Maximum: {numeric_data.max().max():.2f}")
        print(f"    Mean: {numeric_data.mean().mean():.3f}")
        print(f"    Standard deviation: {numeric_data.std().mean():.3f}")

        # Check if features need scaling
        ranges = numeric_data.max() - numeric_data.min()
        max_range = ranges.max()
        min_range = ranges.min()
        if max_range / min_range > 100:
            print(f"  ⚠️  Feature scaling recommended (range ratio: {max_range/min_range:.0f}:1)")
        else:
            print(f"  ✓ Feature ranges are reasonable")

    return categorical_cols, numerical_cols

# Auto-detect dataset structure
column_names = auto_detect_structure('wdbc.data')

print(f"\n{'='*60}")
print("LOADING DATASET WITH AUTO-DETECTED STRUCTURE")
print(f"{'='*60}")

# Load the data with auto-detected column names
try:
    data = pd.read_csv('wdbc.data', header=None, names=column_names)
    print(f"✓ Successfully loaded wdbc.data with auto-detected structure")
except FileNotFoundError:
    print("❌ Error: wdbc.data file not found!")
    print("Please ensure wdbc.data is in the same directory as this script.")
    exit()

print(f"✓ Dataset loaded: {data.shape[0]} samples × {data.shape[1]} columns")

# Display first few rows
print("\nFirst 5 rows with auto-generated column names:")
print(data.head())

# Automatically analyze the dataset
categorical_cols, numerical_cols = analyze_dataset_automatically(data)

# Remove ID column (first column) as it's not needed for prediction
if 'id' in data.columns:
    data_clean = data.drop(['id'], axis=1).copy()
    print(f"\n✓ Removed ID column. New shape: {data_clean.shape}")
else:
    data_clean = data.copy()
    print(f"\n✓ No ID column detected. Using original data.")

# Automatic feature categorization
feature_cols = [col for col in data_clean.columns if col != 'diagnosis']
print(f"\nFeature analysis:")
print(f"  Total features: {len(feature_cols)}")

# Categorize features by suffix if they follow the pattern
mean_features = [col for col in feature_cols if col.endswith('_mean')]
se_features = [col for col in feature_cols if col.endswith('_se')]
worst_features = [col for col in feature_cols if col.endswith('_worst')]

if mean_features and se_features and worst_features:
    print(f"  Mean features: {len(mean_features)}")
    print(f"  Standard Error features: {len(se_features)}")
    print(f"  Worst features: {len(worst_features)}")

    # Use mean features for visualization
    viz_features = mean_features[:6]
    print(f"  Selected for visualization: {viz_features}")
else:
    # Use first few features for visualization
    viz_features = feature_cols[:6]
    print(f"  Selected for visualization: {viz_features}")

# Visualization of key features
if 'diagnosis' in data_clean.columns:
    # Separate by diagnosis
    malignant_data = data_clean[data_clean['diagnosis'] == 'M']
    benign_data = data_clean[data_clean['diagnosis'] == 'B']

    print(f"\nData separation for visualization:")
    print(f"  Malignant samples: {len(malignant_data)}")
    print(f"  Benign samples: {len(benign_data)}")

    # Create feature distribution plots
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle('Feature Distributions by Diagnosis\n(Auto-detected from data)',
                 fontsize=16, fontweight='bold')

    for i, feature in enumerate(viz_features):
        row = i // 3
        col = i % 3

        axes[row, col].hist(benign_data[feature], alpha=0.6, label='Benign',
                           color='blue', bins=25, density=True)
        axes[row, col].hist(malignant_data[feature], alpha=0.6, label='Malignant',
                           color='red', bins=25, density=True)

        # Clean feature name for display
        display_name = feature.replace('_', ' ').title()
        axes[row, col].set_title(f'{display_name}', fontweight='bold')
        axes[row, col].set_xlabel(display_name)
        axes[row, col].set_ylabel('Density')
        axes[row, col].legend()
        axes[row, col].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig("auto_detected_features.png", dpi=300, bbox_inches='tight')
    plt.show()

    # Scatter plot using first two features
    if len(viz_features) >= 2:
        plt.figure(figsize=(12, 8))
        plt.title("Tumor Classification Visualization\n(Auto-detected features)",
                 fontsize=16, fontweight='bold')

        x_feature = viz_features[0]
        y_feature = viz_features[1]

        plt.xlabel(x_feature.replace('_', ' ').title(), fontsize=12)
        plt.ylabel(y_feature.replace('_', ' ').title(), fontsize=12)

        plt.scatter(malignant_data[x_feature], malignant_data[y_feature],
                   color="red", label="Malignant", alpha=0.6, s=50, edgecolors='darkred')
        plt.scatter(benign_data[x_feature], benign_data[y_feature],
                   color="blue", label="Benign", alpha=0.6, s=50, edgecolors='darkblue')

        plt.legend(fontsize=12)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.savefig("auto_detected_scatter.png", dpi=300, bbox_inches='tight')
        plt.show()

# Prepare data for machine learning
print(f"\n{'='*60}")
print("PREPARING DATA FOR KNN ANALYSIS")
print(f"{'='*60}")

# Convert diagnosis to numerical values if it exists
if 'diagnosis' in data_clean.columns:
    unique_diagnosis = data_clean['diagnosis'].unique()
    print(f"Original diagnosis values: {sorted(unique_diagnosis)}")

    # Map diagnosis to numerical values
    if set(unique_diagnosis) == {'M', 'B'}:
        data_clean['diagnosis'] = data_clean['diagnosis'].map({'M': 1, 'B': 0})
        print("✓ Mapped: M=1 (Malignant), B=0 (Benign)")
    else:
        # Handle other possible formats
        diagnosis_mapping = {val: i for i, val in enumerate(sorted(unique_diagnosis))}
        data_clean['diagnosis'] = data_clean['diagnosis'].map(diagnosis_mapping)
        print(f"✓ Mapped diagnosis values: {diagnosis_mapping}")

    # Separate features and target
    X = data_clean.drop(['diagnosis'], axis=1)
    y = data_clean['diagnosis'].values

    print(f"Features (X) shape: {X.shape}")
    print(f"Target (y) shape: {y.shape}")

else:
    print("❌ No diagnosis column found. Cannot proceed with supervised learning.")
    exit()

# Feature scaling
print(f"\nApplying feature scaling...")
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print(f"✓ Features scaled to range [{X_scaled.min().min():.1f}, {X_scaled.max().max():.1f}]")

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)

print(f"\nDataset split:")
print(f"  Training: {X_train.shape[0]} samples ({X_train.shape[0]/len(data_clean)*100:.1f}%)")
print(f"  Testing: {X_test.shape[0]} samples ({X_test.shape[0]/len(data_clean)*100:.1f}%)")

# KNN with different K values
print(f"\n{'='*60}")
print("KNN ANALYSIS WITH DIFFERENT K VALUES")
print(f"{'='*60}")

k_range = np.arange(1, 21)
train_scores = []
test_scores = []

print(f"Testing K values from {k_range[0]} to {k_range[-1]}...")
print(f"{'K':<3} {'Train Acc':<12} {'Test Acc':<12} {'Difference':<12}")
print("-" * 50)

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    train_acc = knn.score(X_train, y_train)
    test_acc = knn.score(X_test, y_test)

    train_scores.append(train_acc)
    test_scores.append(test_acc)

    print(f"{k:<3} {train_acc:<12.4f} {test_acc:<12.4f} {train_acc-test_acc:<12.4f}")

# Find optimal K
optimal_k = k_range[np.argmax(test_scores)]
best_accuracy = max(test_scores)

print(f"\n✓ Optimal K: {optimal_k}")
print(f"✓ Best test accuracy: {best_accuracy:.4f} ({best_accuracy*100:.1f}%)")

# Plot results
plt.figure(figsize=(12, 8))
plt.plot(k_range, train_scores, 'bo-', label='Training Accuracy', linewidth=2)
plt.plot(k_range, test_scores, 'ro-', label='Testing Accuracy', linewidth=2)
plt.axvline(x=optimal_k, color='green', linestyle='--', alpha=0.7,
           label=f'Optimal K={optimal_k}')

plt.title('KNN Performance Analysis\n(Auto-detected dataset structure)',
         fontsize=16, fontweight='bold')
plt.xlabel('Number of Neighbors (K)', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.xticks(k_range)
plt.tight_layout()
plt.savefig("knn_auto_analysis.png", dpi=300, bbox_inches='tight')
plt.show()

# Final model evaluation
print(f"\n{'='*60}")
print("FINAL MODEL EVALUATION")
print(f"{'='*60}")

final_knn = KNeighborsClassifier(n_neighbors=optimal_k)
final_knn.fit(X_train, y_train)
y_pred = final_knn.predict(X_test)

print(f"Final Model: KNN with K={optimal_k}")
print(f"Training Accuracy: {final_knn.score(X_train, y_train):.4f}")
print(f"Testing Accuracy: {accuracy_score(y_test, y_pred):.4f}")

print(f"\nClassification Report:")
target_names = ['Benign', 'Malignant'] if set(data['diagnosis'].unique()) == {'M', 'B'} else None
print(classification_report(y_test, y_pred, target_names=target_names))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
           xticklabels=target_names or ['Class 0', 'Class 1'],
           yticklabels=target_names or ['Class 0', 'Class 1'])
plt.title(f'Confusion Matrix (K={optimal_k})\nAuto-detected Dataset',
         fontsize=14, fontweight='bold')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.savefig("confusion_matrix_auto.png", dpi=300, bbox_inches='tight')
plt.show()

# Summary
print(f"\n{'='*60}")
print("EXPERIMENT SUMMARY")
print(f"{'='*60}")
print(f"✓ Auto-detected {data.shape[1]} columns from data file")
print(f"✓ Identified {len(feature_cols)} features automatically")
print(f"✓ Processed {len(data_clean)} samples successfully")
print(f"✓ Found optimal K={optimal_k} with {best_accuracy*100:.1f}% accuracy")
print(f"✓ No external files needed - everything inferred from data")

overfitting = train_scores[optimal_k-1] - test_scores[optimal_k-1] > 0.05
print(f"✓ Model shows {'overfitting tendency' if overfitting else 'good generalization'}")



```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
from sklearn import tree

def auto_detect_structure(filename='wdbc.data'):
    """Auto-detect dataset structure and generate column names"""
    print("Auto-detecting dataset structure...")

    temp_data = pd.read_csv(filename, header=None)
    print(f"Dataset shape detected: {temp_data.shape}")

    # Generate column names
    num_cols = temp_data.shape[1]
    column_names = ['id', 'diagnosis']

    # For WDBC: 30 features (10 base features × 3 statistics)
    remaining_cols = num_cols - 2

    if remaining_cols == 30:
        print("Detected WDBC format: 30 features")
        base_features = ['radius', 'texture', 'perimeter', 'area', 'smoothness',
                        'compactness', 'concavity', 'concave_points', 'symmetry', 'fractal_dimension']
        statistics = ['mean', 'se', 'worst']

        for stat in statistics:
            for feature in base_features:
                column_names.append(f'{feature}_{stat}')
    else:
        for i in range(remaining_cols):
            column_names.append(f'feature_{i+1}')

    return column_names

# Load and prepare data
print("="*60)
print("DECISION TREE CLASSIFICATION FOR BREAST CANCER DIAGNOSIS")
print("="*60)

# Auto-detect structure and load data
column_names = auto_detect_structure('wdbc.data')

try:
    data = pd.read_csv('wdbc.data', header=None, names=column_names)
    print(f"✓ Successfully loaded dataset: {data.shape[0]} samples × {data.shape[1]} columns")
except FileNotFoundError:
    print("❌ Error: wdbc.data file not found!")
    exit()

# Display basic information
print(f"\nDataset Overview:")
print(f"• Shape: {data.shape}")
print(f"• Memory usage: {data.memory_usage(deep=True).sum() / 1024:.1f} KB")
print(f"• Missing values: {data.isnull().sum().sum()}")

print("\nFirst 5 rows:")
print(data.head())

# Data cleaning
print(f"\n{'='*50}")
print("DATA CLEANING AND PREPARATION")
print(f"{'='*50}")

# Remove null values and ID column
data_clean = data.dropna(axis=1).drop(['id'], axis=1).copy()
print(f"✓ Cleaned data shape: {data_clean.shape}")

# Analyze target variable
print(f"\nTarget Variable Analysis:")
diagnosis_counts = data_clean['diagnosis'].value_counts()
print(f"Diagnosis distribution: {dict(diagnosis_counts)}")

total = len(data_clean)
for diagnosis, count in diagnosis_counts.items():
    name = "Malignant" if diagnosis == "M" else "Benign"
    print(f"• {name} ({diagnosis}): {count} samples ({count/total*100:.1f}%)")

# Separate data for visualization
malignant_data = data_clean[data_clean['diagnosis'] == 'M']
benign_data = data_clean[data_clean['diagnosis'] == 'B']

print(f"\nData separation:")
print(f"• Malignant cases: {len(malignant_data)}")
print(f"• Benign cases: {len(benign_data)}")
print(f"• Class balance ratio: {len(benign_data)/len(malignant_data):.2f}:1")

# Visualization
print(f"\n{'='*50}")
print("DATA VISUALIZATION")
print(f"{'='*50}")

# Feature scatter plot
plt.figure(figsize=(12, 8))
plt.title("Breast Cancer Dataset: Benign vs Malignant Tumors", fontsize=16, fontweight='bold')
plt.xlabel("Radius Mean", fontsize=12)
plt.ylabel("Texture Mean", fontsize=12)

plt.scatter(malignant_data['radius_mean'], malignant_data['texture_mean'],
           color="red", label="Malignant", alpha=0.6, s=50, edgecolors='darkred')
plt.scatter(benign_data['radius_mean'], benign_data['texture_mean'],
           color="blue", label="Benign", alpha=0.6, s=50, edgecolors='darkblue')

plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("decision_tree_scatter.png", dpi=300, bbox_inches='tight')
plt.show()

# Feature distribution analysis
mean_features = [col for col in data_clean.columns if col.endswith('_mean') and col != 'diagnosis'][:6]

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Feature Distributions by Diagnosis\n(Decision Tree Analysis)',
             fontsize=16, fontweight='bold')

for i, feature in enumerate(mean_features):
    row, col = i // 3, i % 3

    axes[row, col].hist(benign_data[feature], alpha=0.6, label='Benign',
                       color='blue', bins=25, density=True)
    axes[row, col].hist(malignant_data[feature], alpha=0.6, label='Malignant',
                       color='red', bins=25, density=True)

    display_name = feature.replace('_mean', '').replace('_', ' ').title()
    axes[row, col].set_title(f'{display_name}', fontweight='bold')
    axes[row, col].set_xlabel(display_name)
    axes[row, col].set_ylabel('Density')
    axes[row, col].legend()
    axes[row, col].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("decision_tree_features.png", dpi=300, bbox_inches='tight')
plt.show()

# Prepare data for machine learning
print(f"\n{'='*50}")
print("PREPARING DATA FOR DECISION TREE")
print(f"{'='*50}")

# Convert diagnosis to numerical
data_clean['diagnosis'] = data_clean['diagnosis'].map({'M': 1, 'B': 0})
print("✓ Converted diagnosis: M=1 (Malignant), B=0 (Benign)")

# Separate features and target
X = data_clean.drop(['diagnosis'], axis=1)
y = data_clean['diagnosis'].values

print(f"Features (X) shape: {X.shape}")
print(f"Target (y) shape: {y.shape}")
print(f"Class distribution: Benign={np.sum(y==0)}, Malignant={np.sum(y==1)}")

# Feature scaling (Min-Max normalization)
print(f"\nApplying Min-Max normalization...")
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print(f"✓ Features normalized to range [{X_scaled.min().min():.1f}, {X_scaled.max().max():.1f}]")

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)

print(f"\nDataset split (70-30):")
print(f"• Training: {X_train.shape[0]} samples")
print(f"• Testing: {X_test.shape[0]} samples")
print(f"• Training class distribution: Benign={np.sum(y_train==0)}, Malignant={np.sum(y_train==1)}")
print(f"• Testing class distribution: Benign={np.sum(y_test==0)}, Malignant={np.sum(y_test==1)}")

# Decision Tree Analysis with Different Parameters
print(f"\n{'='*60}")
print("DECISION TREE HYPERPARAMETER ANALYSIS")
print(f"{'='*60}")

# Test different max_depth values
depths = range(1, 21)
train_accuracies = []
test_accuracies = []

print("Testing different tree depths...")
print(f"{'Depth':<6} {'Train Acc':<12} {'Test Acc':<12} {'Difference':<12} {'Status'}")
print("-" * 60)

for depth in depths:
    dt = DecisionTreeClassifier(max_depth=depth, random_state=42)
    dt.fit(X_train, y_train)

    train_acc = dt.score(X_train, y_train)
    test_acc = dt.score(X_test, y_test)

    train_accuracies.append(train_acc)
    test_accuracies.append(test_acc)

    diff = train_acc - test_acc
    status = "Overfitting" if diff > 0.1 else "Good" if diff < 0.05 else "Slight Overfit"

    print(f"{depth:<6} {train_acc:<12.4f} {test_acc:<12.4f} {diff:<12.4f} {status}")

# Find optimal depth
optimal_depth = depths[np.argmax(test_accuracies)]
best_test_acc = max(test_accuracies)

print(f"\n✓ Optimal max_depth: {optimal_depth}")
print(f"✓ Best test accuracy: {best_test_acc:.4f} ({best_test_acc*100:.1f}%)")

# Plot depth analysis
plt.figure(figsize=(12, 8))
plt.plot(depths, train_accuracies, 'bo-', label='Training Accuracy', linewidth=2)
plt.plot(depths, test_accuracies, 'ro-', label='Testing Accuracy', linewidth=2)
plt.axvline(x=optimal_depth, color='green', linestyle='--', alpha=0.7,
           label=f'Optimal Depth={optimal_depth}')

plt.title('Decision Tree Performance vs Max Depth', fontsize=16, fontweight='bold')
plt.xlabel('Maximum Depth', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.xticks(range(1, 21, 2))
plt.tight_layout()
plt.savefig("decision_tree_depth_analysis.png", dpi=300, bbox_inches='tight')
plt.show()

# Grid Search for best parameters
print(f"\n{'='*50}")
print("GRID SEARCH FOR OPTIMAL PARAMETERS")
print(f"{'='*50}")

param_grid = {
    'max_depth': [3, 5, 7, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}

print("Performing Grid Search...")
grid_search = GridSearchCV(DecisionTreeClassifier(random_state=42),
                          param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

print(f"✓ Best parameters found:")
for param, value in grid_search.best_params_.items():
    print(f"  {param}: {value}")
print(f"✓ Best cross-validation score: {grid_search.best_score_:.4f}")

# Train final model with best parameters
print(f"\n{'='*50}")
print("FINAL DECISION TREE MODEL")
print(f"{'='*50}")

final_dt = grid_search.best_estimator_
final_dt.fit(X_train, y_train)

# Make predictions
y_pred_train = final_dt.predict(X_train)
y_pred_test = final_dt.predict(X_test)

# Calculate accuracies
train_accuracy = accuracy_score(y_train, y_pred_train)
test_accuracy = accuracy_score(y_test, y_pred_test)

print(f"Final Model Performance:")
print(f"• Training Accuracy: {train_accuracy:.4f} ({train_accuracy*100:.1f}%)")
print(f"• Testing Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.1f}%)")
print(f"• Overfitting Check: {train_accuracy - test_accuracy:.4f}")

# Feature Importance Analysis
print(f"\n{'='*50}")
print("FEATURE IMPORTANCE ANALYSIS")
print(f"{'='*50}")

feature_importance = pd.DataFrame({
    'feature': X.columns,
    'importance': final_dt.feature_importances_
}).sort_values('importance', ascending=False)

print("Top 10 Most Important Features:")
print(feature_importance.head(10).to_string(index=False))

# Plot feature importance
plt.figure(figsize=(12, 8))
top_features = feature_importance.head(15)
plt.barh(range(len(top_features)), top_features['importance'])
plt.yticks(range(len(top_features)), top_features['feature'])
plt.xlabel('Feature Importance')
plt.title('Top 15 Feature Importances in Decision Tree', fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig("decision_tree_feature_importance.png", dpi=300, bbox_inches='tight')
plt.show()

# Decision Tree Visualization
print(f"\n{'='*50}")
print("DECISION TREE VISUALIZATION")
print(f"{'='*50}")

# Create a simpler tree for visualization (max_depth=3)
simple_dt = DecisionTreeClassifier(max_depth=3, random_state=42)
simple_dt.fit(X_train, y_train)

plt.figure(figsize=(20, 12))
plot_tree(simple_dt,
          feature_names=X.columns,
          class_names=['Benign', 'Malignant'],
          filled=True,
          rounded=True,
          fontsize=10)
plt.title('Decision Tree Structure (Simplified: max_depth=3)', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig("decision_tree_structure.png", dpi=300, bbox_inches='tight')
plt.show()

# Text representation of tree rules
print(f"\nDecision Tree Rules (Top 3 levels):")
tree_rules = export_text(simple_dt, feature_names=list(X.columns))
print(tree_rules[:1000] + "..." if len(tree_rules) > 1000 else tree_rules)

# Model Evaluation
print(f"\n{'='*60}")
print("COMPREHENSIVE MODEL EVALUATION")
print(f"{'='*60}")

# Classification Reports
print("Training Set Classification Report:")
print(classification_report(y_train, y_pred_train, target_names=['Benign', 'Malignant']))

print("\nTesting Set Classification Report:")
print(classification_report(y_test, y_pred_test, target_names=['Benign', 'Malignant']))

# Confusion Matrices
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Training confusion matrix
cm_train = confusion_matrix(y_train, y_pred_train)
sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues', ax=axes[0],
           xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
axes[0].set_title('Training Set Confusion Matrix', fontweight='bold')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('Actual')

# Testing confusion matrix
cm_test = confusion_matrix(y_test, y_pred_test)
sns.heatmap(cm_test, annot=True, fmt='d', cmap='Reds', ax=axes[1],
           xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
axes[1].set_title('Testing Set Confusion Matrix', fontweight='bold')
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('Actual')

plt.tight_layout()
plt.savefig("decision_tree_confusion_matrices.png", dpi=300, bbox_inches='tight')
plt.show()

# Detailed metrics
tn, fp, fn, tp = cm_test.ravel()
sensitivity = tp / (tp + fn)  # True Positive Rate
specificity = tn / (tn + fp)  # True Negative Rate
precision = tp / (tp + fp)
f1_score = 2 * (precision * sensitivity) / (precision + sensitivity)

print(f"\nDetailed Performance Metrics (Test Set):")
print(f"• True Positives (Malignant correctly identified): {tp}")
print(f"• True Negatives (Benign correctly identified): {tn}")
print(f"• False Positives (Benign misclassified as Malignant): {fp}")
print(f"• False Negatives (Malignant misclassified as Benign): {fn}")
print(f"• Sensitivity (Recall for Malignant): {sensitivity:.4f}")
print(f"• Specificity (Recall for Benign): {specificity:.4f}")
print(f"• Precision (Malignant): {precision:.4f}")
print(f"• F1-Score: {f1_score:.4f}")

# Cross-validation
cv_scores = cross_val_score(final_dt, X_scaled, y, cv=5)
print(f"\n5-Fold Cross-Validation Results:")
print(f"• CV Scores: {[f'{score:.4f}' for score in cv_scores]}")
print(f"• Mean CV Score: {cv_scores.mean():.4f} (±{cv_scores.std()*2:.4f})")

# Final Summary
print(f"\n{'='*60}")
print("EXPERIMENT SUMMARY")
print(f"{'='*60}")
print(f"✓ Dataset: {len(data_clean)} samples with {X.shape[1]} features")
print(f"✓ Algorithm: Decision Tree Classifier")
print(f"✓ Best Parameters: {grid_search.best_params_}")
print(f"✓ Final Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.1f}%)")
print(f"✓ Cross-Validation Score: {cv_scores.mean():.4f}")
print(f"✓ Most Important Feature: {feature_importance.iloc[0]['feature']} ({feature_importance.iloc[0]['importance']:.4f})")
print(f"✓ Model Complexity: Tree depth = {final_dt.get_depth()}, Leaves = {final_dt.get_n_leaves()}")

# Medical interpretation
print(f"\n{'='*60}")
print("MEDICAL INTERPRETATION")
print(f"{'='*60}")
print("Decision Tree Advantages for Medical Diagnosis:")
print("• Interpretable: Can trace decision path for each prediction")
print("• No black box: Rules are transparent and explainable")
print("• Feature importance: Identifies most critical diagnostic features")
print("• Fast prediction: Efficient for real-time diagnosis")

if sensitivity >= 0.95:
    print("✓ Excellent sensitivity: Rarely misses malignant cases")
elif sensitivity >= 0.90:
    print("✓ Good sensitivity: Acceptable for medical screening")
else:
    print("⚠️  Consider improving sensitivity for medical applications")

if specificity >= 0.95:
    print("✓ Excellent specificity: Rarely misclassifies benign cases")
elif specificity >= 0.90:
    print("✓ Good specificity: Acceptable false positive rate")
else:
    print("⚠️  High false positive rate may cause unnecessary anxiety")

print(f"\nTop 3 Decision Factors:")
for i in range(min(3, len(feature_importance))):
    feature = feature_importance.iloc[i]
    print(f"{i+1}. {feature['feature'].replace('_', ' ').title()}: {feature['importance']:.4f} importance")








```

```python
# First import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier  # Added missing import
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import confusion_matrix, classification_report

def auto_detect_structure(filename='wdbc.data'):
    """Auto-detect dataset structure and generate column names"""
    temp_data = pd.read_csv(filename, header=None)

    # Generate column names for WDBC format
    column_names = ['id', 'diagnosis']
    base_features = ['radius', 'texture', 'perimeter', 'area', 'smoothness',
                    'compactness', 'concavity', 'concave_points', 'symmetry', 'fractal_dimension']
    statistics = ['mean', 'se', 'worst']

    for stat in statistics:
        for feature in base_features:
            column_names.append(f'{feature}_{stat}')

    return column_names

# Load data with auto-detected structure
column_names = auto_detect_structure('wdbc.data')
my_data = pd.read_csv('wdbc.data', header=None, names=column_names)

print("Dataset loaded successfully!")
print(f"Shape: {my_data.shape}")
my_data.head()

my_data.info()

# The line below should display (569, 32)
my_data.shape

# To remove the null values from the dataset
my_data.dropna(axis=1, inplace=True)

# Remove the id column
my_dt = my_data.drop(['id'], axis=1)
my_dt.head(3)

# Separate data by diagnosis
The_M = my_dt[my_dt.diagnosis == "M"]
The_M.head()

The_B = my_dt[my_dt.diagnosis == "B"]
The_B.head(6)

# Plot both diagnosis Benign and Malignant
plt.figure(figsize=(10, 6))
plt.title("Random Forest: Benign Tumor VS Malignant")
plt.xlabel("Radius_Mean")
plt.ylabel("Texture_Mean")
plt.scatter(The_M.radius_mean, The_M.texture_mean, color="red", label="Malignant", alpha=0.6)
plt.scatter(The_B.radius_mean, The_B.texture_mean, color="blue", label="Benign", alpha=0.6)
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig("random_forest_plot.png", facecolor='w', bbox_inches='tight', pad_inches=0.3)
plt.show()

# Convert diagnosis to numerical values
my_dt.diagnosis = [1 if i == "M" else 0 for i in my_dt.diagnosis]

# Separate features and target
x_val = my_dt.drop(["diagnosis"], axis=1)
y_val = my_dt.diagnosis.values

# Min-Max Normalization
scaler = MinMaxScaler()
my_x = scaler.fit_transform(x_val)
my_x = pd.DataFrame(my_x, columns=x_val.columns)

print(f"Features shape: {my_x.shape}")
print(f"Target shape: {y_val.shape}")

# Split the dataset into training and testing sets (60% training, 40% testing)
x_train, x_test, y_train, y_test = train_test_split(my_x, y_val, test_size=0.4, random_state=41)

print(f"Training set: {x_train.shape[0]} samples")
print(f"Testing set: {x_test.shape[0]} samples")

# Now let's implement the Random Forest classifier
my_rf = RandomForestClassifier(n_estimators=100, random_state=42)
my_rf.fit(x_train, y_train)

print("Random Forest model trained successfully!")

# Get predictions for both training and testing sets
y_train_pred = my_rf.predict(x_train)
y_test_pred = my_rf.predict(x_test)

# The score() is used to display the classifier accuracy
train_accuracy = my_rf.score(x_train, y_train)
test_accuracy = my_rf.score(x_test, y_test)

print(f"\nRandom Forest Performance:")
print(f"Training Accuracy: {train_accuracy:.4f} ({train_accuracy*100:.1f}%)")
print(f"Testing Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.1f}%)")

# Training confusion matrix (FIXED)
print("\nTraining Confusion Matrix:")
cm_train = confusion_matrix(y_train, y_train_pred)  # Fixed: actual vs predicted
print(cm_train)

# Training classification report (FIXED)
print("\nClassification Report - Training:")
print(classification_report(y_train, y_train_pred, target_names=['Benign', 'Malignant']))

# Testing confusion matrix (FIXED)
print("\nTesting Confusion Matrix:")
cm_test = confusion_matrix(y_test, y_test_pred)  # Fixed: actual vs predicted
print(cm_test)

# Testing classification report (FIXED)
print("\nClassification Report - Testing:")
print(classification_report(y_test, y_test_pred, target_names=['Benign', 'Malignant']))
# Feature importance (bonus - simple addition)
print(f"\nTop 10 Most Important Features:")
feature_importance = pd.DataFrame({
    'feature': x_val.columns,
    'importance': my_rf.feature_importances_
}).sort_values('importance', ascending=False)

print(feature_importance.head(10).to_string(index=False))

# Simple visualization of results
plt.figure(figsize=(12, 5))

# Plot 1: Confusion matrix for test set
plt.subplot(1, 2, 1)
plt.imshow(cm_test, interpolation='nearest', cmap='Blues')
plt.title('Test Set Confusion Matrix')
plt.colorbar()
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.xticks([0, 1], ['Benign', 'Malignant'])
plt.yticks([0, 1], ['Benign', 'Malignant'])

# Add text annotations
for i in range(2):
    for j in range(2):
        plt.text(j, i, cm_test[i, j], ha='center', va='center', fontsize=14, fontweight='bold')

# Plot 2: Feature importance
plt.subplot(1, 2, 2)
top_features = feature_importance.head(10)
plt.barh(range(len(top_features)), top_features['importance'])
plt.yticks(range(len(top_features)), top_features['feature'])
plt.xlabel('Importance')
plt.title('Top 10 Feature Importances')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig("random_forest_results.png", dpi=300, bbox_inches='tight')
plt.show()

# Summary
print(f"\n{'='*50}")
print("EXPERIMENT SUMMARY")
print(f"{'='*50}")
print(f"✓ Dataset: {len(my_dt)} samples with {x_val.shape[1]} features")
print(f"✓ Algorithm: Random Forest with {my_rf.n_estimators} trees")
print(f"✓ Training Accuracy: {train_accuracy:.4f}")
print(f"✓ Testing Accuracy: {test_accuracy:.4f}")
print(f"✓ Most Important Feature: {feature_importance.iloc[0]['feature']}")

if test_accuracy > 0.95:
    print("✓ Excellent performance achieved!")
elif test_accuracy > 0.90:
    print("✓ Good performance achieved!")
else:
    print("⚠️  Consider tuning parameters for better performance")
```
